```python
import gym
import numpy as np
#from gym.envs.registration import register
#register(
#    id='FrozenLakeNotSlippery-v0',
#    entry_point='gym.envs.toy_text:FrozenLakeEnv',
#    kwargs={'map_name' : '4x4', 'is_slippery': False},
#    max_episode_steps=100,
#    reward_threshold=0.78, # optimum = .8196
#)
#env = gym.make('FrozenLakeNotSlippery-v0')
env = gym.make('FrozenLake-v0')
#env = gym.make('FrozenLake8x8-v0')
#env = gym.make('Taxi-v2')
```

```python
def play(policy,n_plays = 1000, max_steps = 1000, gamma = 1.0):
    total_reward = 0
    for x in range(n_plays):
        state = env.reset()
        for step in range(max_steps):
            state, reward, done, info = env.step(policy[state])
            total_reward += (gamma ** step) * reward
            if done:
                break
    return total_reward/n_plays

def sample_play(policy):
    done = False
    state = env.reset()
    total_reward = 0
    while done == False:
        env.render()
        state, reward, done, info = env.step(policy[state])
        total_reward += reward
    return total_reward

def evaluate_policy(policy, gamma, theta = 0.0001):
    V = np.zeros(env.env.nS)
    while True:
        delta = 0
        for s in range(env.env.nS):
            state_value = 0
            for prob, next_state, reward, done in env.env.P[s][policy[s]]:
                state_value += prob*(reward + gamma * V[next_state])
            delta = max(delta, np.abs(V[s] - state_value))
            V[s] = state_value
        if delta < theta:
            break
    return V

def policy_iteration(gamma = 0.9):
    policy = np.random.randint(env.env.nA, size=env.env.nS)
    
    while True:
        
        change = False
        V = evaluate_policy(policy, gamma)
        
        for s in range(env.env.nS):
            
            chosen_a = policy[s]
            actions = np.zeros(env.env.nA)
            
            for a in range(env.env.nA):
                
                action_value = 0
                for prob, next_state, reward, done in env.env.P[s][a]:
                    action_value += prob*(reward + gamma * V[next_state])
                    
                actions[a] = action_value
            best_a = np.argmax(actions)
            policy[s] = best_a
            if chosen_a != best_a:
                change=True
        if change == False:
            return policy, V
```

```python
policy, V = policy_iteration(gamma = 0.999)
acc = play(policy)
print(acc)
#print(policy.reshape((8,8)))
sample_play(policy)
```
